```python
import numpy as np
import helpers as h
from matplotlib import pyplot as plt

lamGrid = np.linspace(380,420,320)

rvcc=0
c=30000
nLam = len(lamGrid)
windows = np.zeros((nLam,4),dtype=np.float32)
z = 1. +rvcc/c
#brown
#make window functions

#this creates a triangle around h.cahLam
#by making 
d0 = abs(lamGrid-h.cahLam*z)/h.lineWid

s = (d0<=1.0).nonzero()

print(lamGrid[s])
if len(s) > 0:#should always be true
    windows[s,0]=1.-d0[s] 
print(s)

plt.figure()
print(windows[s,0])
plt.plot(lamGrid[s],windows[s,0][0],'bo')
plt.show()
plt.close()

d1 = abs(lamGrid-h.cakLam*z)/h.lineWid
s = (d1<=1.0).nonzero()
if len(s) > 0:
    windows[s,1]=1.-d1[s] 

d2 = abs(lamGrid-h.lamR*z)*2./h.conWid
s = (d2<=1.0).nonzero()
if len(s) > 0:
    windows[s,2]=1.

#v-band inclusion
d3 = abs(lamGrid-h.lamB*z)*2./h.conWid
s = (d3<=1.0).nonzero()
if len(s) > 0:
    windows[s,3]=1.
```
